```python
"""
for in 반복문
자료형에 담긴 자료들을 하나씩 꺼내고 싶을때 사용한다.
이렇게 반복문을 활용해서 실행을 하게 되면 컬럼이 아무리 많아도
반복문을 통해서 한번에 인코딩 할 수 있다.
"""
# 활용할 데이터 선택 - 원본 데이터 불러오기
df = data[['가격', '호수', '칼로리', '원산지', '살찔까요']].copy()
df

# 한 번에 레이블 인코딩
from sklearn.preprocessing import LabelEncoder

for col in cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

# 인코딩 결과 확인
df
```

## **원핫(one-hot) 인코딩**

```python
# 활용할 데이터 선택
df = data[['가격', '호수', '칼로리', '원산지', '살찔까요']].copy()
df
```

```python
# 원핫인코딩
# 데이터프레임을 입력받기 때문에 대괄호를 2번 써야 합니다. (df[['원산지']])
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output=False)   # 기존 sparse=False 대신 수정, 사이킷런 버전 문제
cat = ohe.fit_transform(df[['원산지']])
cat
```

```python
# 데이터 프레임 합치기
df = pd.concat([df,df_cat],axis=1)
df
```

```python
# 기존 컬럼 삭제
df = df.drop(['원산지'], axis=1)
df
```

## **당뇨병 데이터**

```python
# 당뇨병 데이터 불러오기
from sklearn.datasets import load_diabetes
dataset_2=load_diabetes()
dataset_2
```

```python
# 데이터프레임 형태로 변환
pd.DataFrame(data=dataset_2.data, columns=dataset_2.feature_names)
diabetes_df = pd.DataFrame(data=dataset_2.data, columns=dataset_2.feature_names)
diabetes_df
```

```python
# 타겟 추가하기
diabetes_df['target'] = dataset_2.target
diabetes_df.head()
```

# **2-6. 머신러닝**

## **의사결정나무 (DecisionTree)**

**학습(tr)/검증(te) 데이터 분리**

```python
# 모듈 불러오기
from sklearn.model_selection import train_test_split
# X_train, X_test // y_train, y_test 분리
X_train, X_test, y_train, y_test = train_test_split(
    cancer_df.drop('target', axis=1), cancer_df['target'], test_size=0.3, random_state=1004)
```

```python
# train data 확인
X_train.head()
```

```python
y_train.head()
```

```python
# 데이터 크기(분류 비율) 확인
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

**분리데이터|를 머신러닝 모델에 학습**

```python
# 모델 호출 후 변수지정
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()

# 학습
model.fit(X_train, y_train)

# 예측
pred = model.predict(X_test)
pred
```

**모델평가**

```python
# 정확도 accuracy_score(실제값, 예측값)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

## **선형회귀 (LinearRegression)**

**학습(tr)/검증(te) 데이터 분리**

```python
# 모듈 불러오기
from sklearn.model_selection import train_test_split
# X_train, X_test // y_train, y_test 분리
X_train, X_test, y_train, y_test = train_test_split(
    diabetes_df.drop('target', axis=1), diabetes_df['target'], test_size=0.3, random_state=1004)
```

```python
# train 데이터 확인
X_train.head()
```

```python
y_train.head()
```

```python
# 데이터 크기(분류 비율) 확인
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

**분리데이터|를 머신러닝 모델에 학습**

```python
# 모델 호출 후 변수지정
from sklearn.linear_model import LinearRegression
model = LinearRegression()

# 학습
model.fit(X_train, y_train)

# 예측
pred = model.predict(X_test)
pred
```

**평가 (Mean Squared Error,MSE)**
- 회귀모델에서는 MSE 를 성능지표로 많이 사용

```python
# 정확도 mean_squered_error(실제값, 예측값)
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, pred)
```
